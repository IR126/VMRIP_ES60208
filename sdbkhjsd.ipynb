```python
import pandas as pd
import numpy as np

soc = pd.read_parquet(r"C:\Users\ACER\OneDrive\Desktop\6th sem\RBPM\Project\battery-soc\bsoc\results\recommissioned_batteries\battery02\soc.parquet")
ocv = pd.read_csv(r"C:\Users\ACER\OneDrive\Desktop\6th sem\RBPM\Project\battery-soc\bsoc\results\recommissioned_batteries\battery02\ocv_lookup.csv")

cyc0 = soc[soc["cycle_num"] == 0].sort_values("time").reset_index(drop=True)

ocv_fn = lambda s: np.interp(s, ocv["SOC"], ocv["OCV_V"])

# R0 = (OCV - V_terminal) / I  at start of discharge
I = 2.5
v_terminal = cyc0["voltage"].iloc[5]   # skip row 0 which may be noisy
soc_val    = cyc0["SOC"].iloc[5]
ocv_val    = ocv_fn(soc_val)

R0_estimate = (ocv_val - v_terminal) / I
print(f"OCV={ocv_val:.4f}  V_terminal={v_terminal:.4f}  R0_estimate={R0_estimate:.6f} Ω")
```
